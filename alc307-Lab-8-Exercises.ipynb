#### **Exercise 2**

Calculate the average item price "the hard way," by summing the values in the column and dividing. Just like you did in math class. (Sorry.)

```python
#Make a count
count = 0

for x in range(len(chip)):
    count += float(chip.loc[x, "item_price"])
    
count = count / chip["item_price"].count()
print("Average Price: ${price:.2f}".format(price= count))
```

#### **Exercise 3**

Now, calculate the mean again, using a pandas function. This is the easy way to do the above.

```python
print("Average Price: ${price:.2f}".format(price = float(chip["item_price"].mean())))
```

#### **Exercise 5**

Using a query mask and [Pandas string methods](https://pandas.pydata.org/pandas-docs/stable/reference/series.html#string-handling), create a new dataframe of just the times people ordered chips. Print the first five such times.

```python
#First just get chips orders by creating a query mask
chip_orders = chip["item_name"].str.contains("Chips")

#Apply the mask to a subset of the dataframe
chip_orders_sdf = chip[chip_orders]

#Make a copy of the subset for separate usage
chip_orders_csdf = chip_orders_sdf.copy()

print("======= CHIP ORDERS =======")
chip_orders_csdf.head(5)
```
